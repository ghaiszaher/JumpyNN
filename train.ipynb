```python
import keras
from keras.models import Sequential
from keras.layers import Dense, RNN
```

```python
import warnings
import numpy as np
import keras.backend as K
import json

warnings.filterwarnings('ignore')
```

```python
folder = "./gameplay_logs"
d=f'{folder}/features_X_y.npz'
```

```python
with open(d, "rb") as f:
    data = np.load(f)
    X = data["X"]
    y = data["y"]

print(X.shape)
print(y.shape)
```

```python
def absolute_diff(yTrue,yPred):
    return K.mean(K.abs(K.round(yTrue) - K.round(yPred)))


classifier = Sequential(name="Test")
classifier.add(Dense(output_dim = 200, init = 'normal', activation = 'relu', input_dim = X.shape[1]))
classifier.add(Dense(output_dim = 100, init = 'normal', activation = 'relu'))
classifier.add(Dense(output_dim = 1, init = 'normal'))

classifier.compile(optimizer = 'adam', loss = 'mean_squared_error', metrics = ['accuracy', absolute_diff])
```

```python
classifier.fit(X,y, batch_size=50, epochs=250)
```

```python
i = np.random.randint(X.shape[0], size=1)


pred = classifier.predict(X[i,:])
np.round(pred), y[i,:]
```

```python
np.mean(np.abs(np.round(classifier.predict(X))-y))
```

```python
y_pred = classifier.predict(X)
l = len(y)
acc = sum([np.round(y_pred[i])==y[i] for i in range(l)])/l
acc
```

```python
loss, acc, rounded_diff = classifier.evaluate(X, y, verbose=0)
print("        loss =", loss)
print("         acc =", acc)
print("rounded_diff =", rounded_diff)
```

```python
classifier.save(f"{folder}/nn_trained.hdf5")


model_config = classifier.get_config()
with open(f"{folder}/nn_model_config.json", "w") as f:
    json.dump(model_config, f)
    
classifier.save_weights(f"{folder}/nn_weights.hdf5")

print("Saved!")
```
